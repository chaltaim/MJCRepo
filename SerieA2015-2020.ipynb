Select strikers and forwards only to see relationship attributes have on goals for forwards.

```python
df1520['Striker'] = df1520['player_positions'].str.contains('ST|CF')
df1520ST = df1520[df1520['Striker'] == True] 
df1520ST.shape
```

Find non-penalty statistics, and add goals per 90 minutes played statistic. Remove strikers who played under 500 minutes for the season.

```python
df1520ST['Glsp'] = df1520ST.loc[:,'Gls'] / df1520ST.loc[:,'90s']
df1520ST['Astp'] = df1520ST.loc[:,'Ast'] / df1520ST.loc[:,'90s']
df1520ST['npGls'] = df1520ST.loc[:,'Gls'] - df1520ST.loc[:,'PK']
df1520ST['npGlsp'] = df1520ST.loc[:,'npGls'] / df1520ST.loc[:,'90s']

df1520ST = df1520ST[df1520ST['Min'] > 500] 
```

Lastly, ensure all attributes are treated as numbers for correlations to work.

```python
col = df1520ST.columns
attributes = col[44:73]

for i in df1520[attributes]: 
    
    df1520ST[i] = df1520ST[i].str[0:2]
    df1520ST[i] = df1520ST[i].astype(float)
```
